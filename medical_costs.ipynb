```python
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
```

```python
mc_df = pd.read_csv("medical_costs.csv")
mc_df.head()
```

```python
mc_df.describe()
```

```python
g = sns.displot(data=mc_df, kind="hist", x="Age", col="Region", hue="Region", bins=[10,20,30,40,50,60,70,80])
# g.set_axis_labels("FOO", "BAR")
```

```python
sns.relplot(data=mc_df, kind="line", x="Age", y="Medical Cost", aspect=2)
```

```python
binned_bmi_df = pd.cut(mc_df["BMI"], bins=10)
mc_df["BMI_binned"] = binned_bmi_df
mc_df["BMI_binned_mid"] = mc_df["BMI_binned"].apply(lambda x: x.mid)
sns.relplot(data=mc_df, kind="line", x="BMI_binned_mid", y="Medical Cost", aspect=2)
```

```python
sns.displot(data=mc_df, x="Medical Cost", hue="Smoker")
```

```python
sns.displot(data=mc_df, x="Medical Cost", hue="BMI_binned_mid", palette="Reds")
```
